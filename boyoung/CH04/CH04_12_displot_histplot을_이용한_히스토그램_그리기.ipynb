# displot_histplot을 이용한 히스토그램 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(15,5)})
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

#distplot

```python
# 데이터를 바로 가져다가 올려줌(자동적으로 데이터를 나눔)
sns.distplot(df['sepal_length']) # 해당 데이터의 밀도 표시
```

```python
# 구간개수 설정
sns.distplot(df['sepal_length'], bins = 15) 
```

#histplot

```python
sns.histplot(df['sepal_length']) # count로 표시
```

```python
sns.histplot(df['sepal_length'], bins =20)
```

```python
sns.histplot(df['sepal_length'], bins =20, kde = True)
```

-

```python
# 카테고리별 분류
sns.histplot(df['sepal_length'], bins =20, kde = True, stat = 'density', hue = 'species')# hue로 나눠야하는데, 데이터 프레임을 할당한게 아니여서 불가함
```

```python
# histpot data사용가능, x축 할당, species기준으로 나눔
sns.histplot(data =df, x = 'sepal_length', bins =20, kde = True, stat = 'density', hue = 'species')
```

```python
# x축, y축 변경
sns.histplot(data =df, y = 'sepal_length', bins =20, kde = True, stat = 'density', hue = 'species')
```

```python
# histplot에 x와 y를 다 넣으면 히트맵이됨
sns.histplot(data =df, y = 'sepal_length', x = 'sepal_width', cbar = True )
```

#displot

```python
sns.displot(df['sepal_length'])
```

```python
sns.displot(data = df, x = 'sepal_length')
```

- 그림 크기 변경

```python
sns.displot(data = df, x = 'sepal_length', height = 5, aspect = 3) # width = height * aspect
```

```python
# histgram 과 한 옵션사용(1)
sns.displot(data = df, x = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species') 
```

```python
# histgram과 한 옵션사용(2)
sns.displot(data = df, y ='sepal_length', x = 'sepal_width', cbar = True) 
```

# jointplot

```python
# 기본 사용법
sns.jointplot(data = df,y ='sepal_length', x = 'sepal_width' ) # 점들 해당 구간에서의 밀도
```

```python
# histplot의 ㅎ여식으로 변경
sns.jointplot(data = df,y ='sepal_length', x = 'sepal_width', kind ='hist', cbar = True ) # 점들 해당 구간에서의 밀도
```

```python
# 그림 크기 변경(정사각형의 형태)
# height 옵션에 숫자 넣어주면 가로세로 정사각형으로 변경됨
sns.jointplot(data = df,y ='sepal_length', x = 'sepal_width', kind ='hist', cbar = True, height = 10) 
```
