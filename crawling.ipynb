```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd


def returnDataFrame(param):
    param_rows = param[0].find_all('tr')
    param_result = []
    for row in param_rows:
        cells = row.find_all('td')
        values = [cell.text.strip() for cell in cells]
        param_result.append(values)
    param_df = pd.DataFrame(param_result)
    return param_df

def find_element_with_refresh(driver, locator, timeout=120):
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located(locator)
        )
        return element
    except TimeoutException:
        print("Timeout while waiting for element. Refreshing page.")
        driver.refresh()
        return find_element_with_refresh(driver, locator, timeout)


list = ['2023-1'
, '2023-2'
, '2023-3'
, '2023-4'
, '2023-5'
, '2023-6'
, '2023-7'
, '2023-8'
, '2023-9'
, '2023-10'
, '2023-11'
, '2023-12'
, '2024-1'
, '2024-2']

result = pd.DataFrame() 
for idx in list:
    # Chrome WebDriver 초기화
    driver = webdriver.Chrome()

    # 페이지 열기
    driver.get("https://www.wunderground.com/history/monthly/us/tn/nashville/KBNA/date/" + idx)

    # 페이지가 로드될 때까지 기다리기 (예: 10초)
    find_element_with_refresh(driver, (By.ID, "inner-content"))
    # 페이지의 내용 가져오기
    page_source = driver.page_source

    # 여기서 BeautifulSoup 또는 다른 방법을 사용하여 페이지를 크롤링할 수 있습니다.
    # 예: BeautifulSoup을 사용한 예제
    from bs4 import BeautifulSoup

    soup = BeautifulSoup(page_source, 'html.parser')
    # 이제 soup을 사용하여 필요한 정보를 추출하거나 원하는 작업을 수행할 수 있습니다.
    # BeautifulSoup을 사용하여 테이블 선택
    time = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(1) > table')
    temperature = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(2) > table')
    dp = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(3) > table')
    hum = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(4) > table')
    wind = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(5) > table')
    pressure = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(6) > table')
    precipitation = soup.select('#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted > table > tbody > tr > td:nth-child(7) > table')


    time_df = returnDataFrame(time)
    temperature_df = returnDataFrame(temperature)
    dp_df = returnDataFrame(dp)
    hum_df = returnDataFrame(hum)
    wind_df = returnDataFrame(wind)
    pressure_df = returnDataFrame(pressure)
    precipitation_df = returnDataFrame(precipitation)

    df = pd.concat([time_df, temperature_df, dp_df, hum_df, wind_df, pressure_df, precipitation_df], axis=1)
    result = pd.concat([result, df], ignore_index=True)

    # WebDriver 종료
    driver.quit()
    
print(result)
```
